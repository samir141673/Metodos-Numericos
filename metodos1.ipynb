```
import numpy as np
import math
import scipy.special

def fx(x,i):
  return pow(x,i)

def cifras_significativas(n):
  return 0.5*pow(10,2-n)

def termino(x,y=1.648721):
  i=0
  f=0
  lista=[]
  f+=fx(x,i)/math.factorial(i)
  lista.append(f)
  print("termino ",(i+1)," resultado ",f)
  Et=abs((y-f)/f)*100
  print("error relativo porcentual ",Et)
  i+=1
  print("===================================")
  cant_cifras=cifras_significativas(6)
  while cant_cifras<Et:
    f+=fx(x,i)/math.factorial(i)
    print("termino ",(i+1)," resultado ",f)
    Et=abs((y-f)/f)*100
    print("error relativo porcentual ",Et)
    Ea=((f-lista[i-1])/f)*100
    print("valor de la estimacion aproximada de error ",Ea)
    print("===============================================")
    lista.append(f)
    i+=1
termino(0.5)
```

```
import numpy as np
import math
import scipy.special

def fx(x,i,j):
  return (pow(x,i)*(-1)**j/math.factorial(i))
  
def cifras_significativas(n):
  return 0.5*pow(10,2-n)


def termino(x):
  ter=1 # nro de terminos
  j=0 #contador de lista
  f=0 #inicializar funcion
  lista=[] #inicializar lista de funciones acumuladas
  f+=fx(x,ter,2)
  lista.append(f)
  j+=1
  Et=abs(0.47-f)/0.47  #valor verdadero le ponemos 0.4794255386041834 =0.4795
  print("termino ",ter," resultado ",f)
  print("error relativo porcentual ",Et)
  print("====================================")
  k=3 #exponente
  i=1 #signo
  cant_cifras=cifras_significativas(6)
  while cant_cifras<Et:
    
    f+=fx(x,k,i)
    lista.append(f)
    Et=abs(0.47-lista[j-1])/0.5
    ter+=1
    print("termino ",ter," resultado ",f)
    print("error relativo porcentual ",Et)
    Ea=((f-lista[j-1])/f)*100
    print("valor de esta estimacion apriximada del error es ", Ea)
    k+=2
    i+=1
    j+=1
    print("=========================================")

termino(0.5)
```

```
import numpy as np
import math
import scipy.special
# x - x^3/3! + x^5/5! - x^7/7! + 6 cifras significativs
def fx(x,i,j):
  return (pow(x,i)*(-1)**j/math.factorial(i))
  
def cifras_significativas(n):
  return 0.5*pow(10,2-n)

def termino(x):
  i=1 # i  exponete de x 
  f=0
  lista=[]
  l=0 #contador de la lista
  f+=fx(x,i,2)
  lista.append(f)
  l+=1
  print("termino ", 1 ," resultado ",f)
  #valor verdader 0.48
  Et=abs(0.48-f/0.48)*100
  print("error relativo porcentual ",Et)
  print("===============================")
  cant_cifras=cifras_significativas(6)
  j=1 #sgino 
  k=3 #exponente
  termino=2
  while cant_cifras<Et:
    
    f+=fx(x,k,j)
    j+=1
    lista.append(f)
    l+=1
    k+=2
    Et=(abs(0.48-lista[l-1])/0.48)*100
    print("termino ",termino," resultado ",f)
    termino+=1
    print("error relativo porcentual ",Et)
    Ea=((f-lista[l-1])/f)*100
    print("valor de esta estimacion apriximada del error es ", Ea)
    
    print("=========================================")

termino(0.5)
```
